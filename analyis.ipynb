### Data used ðŸ’½
The group collected 100 listings in total from the Vicy Market on the dark web.

The focus was to collect data in a objective manner. This was done by choosing random listings listed on the site that were recommended. The listings recommended were most often from sellers with many reviews and sales. These sellers are the good and reliable sellers on the site.

The data is checked for duplicates before generating the graphs. Data collected contains a column called 'Listing ID'. This is a unique identifier for a listing and it generated from the URL in the Excel file. In this Notebook we check whether there are duplicate Listing ID's in the dataset and point out where the duplicates are located so we can exchange them for new, non duplicated listings, to improve the quality of the data and analysis.

### Results generated in this Jupyter Notebook
- Duplicate listings in the dataset
- Top 10 sellers based on amount of listings on Vice City Market
- Top 10 highest priced listings on Vice City Market
- Ratio of legal/illegal/depends listings
- Top 3 categories with the most listings

```python
# Load all dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load and initialize the Excel file
file = 'Black Market analysis - Vice City Market.xlsx'
```

## Prints a table of duplicates when there are duplicates Listing ID's in the dataset

```python
df = pd.read_excel(file, usecols=['Listing ID'])
df[df.duplicated(['Listing ID'], keep=False)]
```

# Top 10 sellers with the most listing on Vicy City Black Market ðŸ“Š

```python
# Top 10 sellers with the most listing on Vicy City Black Market
df = pd.read_excel(file, usecols=['Seller'])
df.value_counts().head(10).plot.bar(color='green')
plt.title('Top 10 sellers based on amount of listings on Vice City Market')
plt.ylabel('Amount of listings')
plt.show()
```

# Top 10 highest priced products ðŸ“Š

```python
# Top 10 highest priced products
df = pd.read_excel(file, usecols=['Title', 'Price'], index_col=0)
df.sort_values('Price', ascending=False).head(10).plot.bar(color='green')
plt.title('Top 10 highest priced listings on Vice City Market')
plt.xlabel(df.columns[0])
plt.ylabel('Price of listing')
plt.show()
```

# Top 3 categories with the most listings ðŸ“Š

```python
# Top 3 categories with the most listings
df = pd.read_excel(file, usecols=['Category'])
df.value_counts().head(10).plot.bar(color='green')
plt.title('Top 3 categories with the most listings')
plt.ylabel('Amount of listings')
plt.show()
```

# Ratio of legal/illegal/depends listings ðŸ“Š

```python
# Ratio of legal/illegal/depends listings
df = pd.read_excel(file, usecols=['Legal?'])
df.value_counts().head(10).plot.bar(color='green')
plt.title('Ratio of legal/illegal/depends listings')
plt.ylabel('Amount of listings')
plt.show()
```
