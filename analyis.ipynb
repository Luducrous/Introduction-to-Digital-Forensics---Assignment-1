```python
# Load all dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load and initialize the Excel file
file = 'BlackMarketFinal.xlsx'
```

## Prints a table of duplicates when there are duplicates Listing ID's in the dataset

```python
df = pd.read_excel(file, usecols=['Listing ID'])
df[df.duplicated(['Listing ID'], keep=False)]
```

As can be seen there are not any duplicate listing ID's in our datasheet.

# Top 10 sellers with the most listing on Vice City Black Market ðŸ“Š

```python
# Top 10 sellers with the most listing on Vicy City Black Market
df = pd.read_excel(file, usecols=['Seller'])
df.value_counts().head(10).plot.bar(color='green')
plt.title('Top 10 sellers based on amount of listings on Vice City Market')
plt.ylabel('Amount of listings')
plt.show()
```

In this graph we displayed the top 10 sellers based on the amount of listings they have within our datasheet. MiamiVice is clearly the biggest, within our 100 products they are the vendor for 18 of them, and they at the same time are also the biggest on the whole market. Trusteelucky is the second biggest with 9 products and Psychonauta02 is the third biggest with 6. These three vendors make up the Top 3 in our data sheet.

# Top 10 highest priced products ðŸ“Š

```python
# Top 10 highest priced products
df = pd.read_excel(file, usecols=['Title', 'Price'], index_col=0)
df.sort_values('Price', ascending=False).head(10).plot.bar(color='green')
plt.title('Top 10 highest priced listings on Vice City Market')
plt.xlabel(df.columns[0])
plt.ylabel('Price of listing')
plt.show()
```

This graph is not particulary for the assignment however we wanted to add this as an extra analysis. We were curious what the top 10 most expensive products would look like. As can be seen from the graph the 1lb of pure crystal meth is the most expensive, this is quite obvious because 1lb is very much especially for a hard drug. We could see that expensive listings on ViceCity were on in most cases in the category of drugs. ViceCity also had alot of 'courses'/'education', for example e-books and video lectures explaining certain (il)legal activities, these listings were in general less than 20 dollars and in most cases under 10 dollars.

# Top 3 categories with the most listings ðŸ“Š

```python
# Top 3 categories with the most listings
df = pd.read_excel(file, usecols=['Category'])
df.value_counts().head(10).plot.bar(color='green')
plt.title('Top 3 categories with the most listings')
plt.ylabel('Amount of listings')
plt.show()
```

We divided all the products within 9 categories, this way we could see which categories come more often than the other. As can be seen from the graph drugs is the most common listing on ViceCity. On a close second place we have the category courses and education. Software is third.

# Ratio of legal/illegal/depends listings ðŸ“Š

```python
# Ratio of legal/illegal/depends listings
df = pd.read_excel(file, usecols=['Legal?'])
df.value_counts().head(10).plot.bar(color='green')
plt.title('Ratio of legal/illegal/depends listings')
plt.ylabel('Amount of listings')
plt.show()
```

For all listings we said if they were illegal, legal or if it depends on the scenario (consent and intention). As can be seen most of the products were illegal, this is because alot of the listings are in the category of drugs and they are illegal in most countries especially the hard drugs that are being sold alot on the ViceCity market. Depends is in a close second place, this is because a lot of products can be used illegaly or legal, it depends on the consent and the intention of the user.
